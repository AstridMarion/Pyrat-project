```python
# External imports
import sys
import os
import pprint

#Â Add needed directories to the path
sys.path.append(os.path.join("..", "players"))

# PyRat imports
from pyrat import Game,GameMode,StartingLocation
from GreedyEachTurn import GreedyEachTurn
```

```python
CONFIG={"nb_cheese":12,
        "mud_percentage":43,
        "random_seed":43,
        "trace_length":1000,
        "game_mode":GameMode.SEQUENTIAL
        }
```

```python
# Instantiate a game with specified arguments
game = Game(**CONFIG)

# Instantiate a player and add it to the game
player1 = GreedyEachTurn("V1")
player2 = GreedyEachTurn("V2")

game.add_player(player1,location=StartingLocation.RANDOM)
game.add_player(player2,location=StartingLocation.RANDOM)


# Start the game and
stats = game.start()
```
