```python
"""l'objectif est de trouver les bon paramètre alpha et beta pour l'arlorithme AntColonies"""


# External imports
import sys
import os
import pprint

# Add needed directories to the path
sys.path.append(os.path.join("..", "players"))

# PyRat imports
from pyrat import Game,StartingLocation
from AntColonies import AntColonies
from tqdm import tqdm

import numpy as np
import matplotlib.pyplot as plt
```

```python
CONFIG = {"mud_percentage": 20.0,
          "cell_percentage": 80.0,
          "wall_percentage": 60.0,
          "maze_width": 25,
          "maze_height": 10,
          "nb_cheese": 41,
          "trace_length":1000,
          "trace_length": 50,
          "random_seed":42,
          "mud_range":[4,9]}
player_position=137
cheese=[89, 140, 41, 142, 17, 60, 122, 2, 63, 87, 206, 241, 179, 188, 237, 62, 155, 225, 3, 205, 19, 190, 220, 209, 164, 150, 65, 9, 145, 53, 84, 16, 92, 245, 199, 54, 104, 43, 208, 165, 243]

```

```python
game = Game(**CONFIG)

player = AntColonies()
maze=game.maze

class game_state:
    def __init__(self):
        self.player_locations = {player.name: player_position}
        self.cheese =cheese
game_state = game_state()
graph,_=player.maze_to_graph(maze,game_state)
```

```python
def ant(alpha,beta,fourmis):
    player.alpha=alpha
    player.beta=beta
    player.ant=fourmis
    try :
        path=player.AntColonies(graph,player_position)
        weight=0
        for i in range(len(path)-1):
            weight+=graph.get_weight(path[i],path[i+1])
    except:
        weight=float('inf')
    return weight
```

```python
#data of the regression
alpha_lenth=6
beta_lenth=6
fourmis=2000
alpha = np.linspace(47, 53, alpha_lenth)
beta = np.linspace(47, 53, beta_lenth)
ALPHA, BETA = np.meshgrid(alpha, beta)

# calcule of ant for every alpha and beta 

Z =np.zeros((alpha_lenth,beta_lenth))
for i in tqdm(range(alpha_lenth),"x"):
    for j in range(beta_lenth):
        print(i,j)
        Z[i,j]=ant(alpha[i],beta[j],fourmis)


# see the result
plt.figure(figsize=(8, 6))
plt.pcolormesh(ALPHA, BETA, Z, shading='auto', cmap='viridis')  
plt.colorbar(label="Valeur")  
plt.title("Dégradé de couleur basé sur les valeurs")
plt.xlabel("alpha")
plt.ylabel("beta")
plt.show()
```
