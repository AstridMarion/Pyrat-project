This script creates a sample game, in which two players compete in the maze.

<h1 style="background-color: gray;
           color: black;
           padding: 20px;
           text-align: center;">IMPORTS</h1>

```python
# External imports
import sys
import os
import pprint

# Add needed directories to the path
sys.path.append(os.path.join("..", "players"))


# PyRat imports

from pyrat import Game, PlayerSkin, StartingLocation, RenderMode, GameMode, Action, RandomMazeAlgorithm
from GreedyEachTurn import GreedyEachTurn
from AntColonies import AntColonies


```

<h1 style="background-color: gray;
           color: black;
           padding: 20px;
           text-align: center;">CONSTANTS</h1>

Let's configure the game with a dictionary.

```python
# Customize the game elements
CONFIG = {"mud_percentage": 20.0,
          "cell_percentage": 80.0,
          "wall_percentage": 60.0,
          "maze_width": 25,
          "maze_height": 10,
          "nb_cheese": 41,
          "trace_length": 50,
          "mud_range":[4,9],
          "game_mode":GameMode.SEQUENTIAL}
```

<h1 style="background-color: gray;
           color: black;
           padding: 20px;
           text-align: center;">RUN THE GAME</h1>

We perform a single game with the configuration defined above.

```python
# Instantiate a game with specified arguments
game = Game(**CONFIG)

# Instantiate players in distinct teams
player_1 = GreedyEachTurn(skin=PlayerSkin.RAT)
player_2 = AntColonies(skin=PlayerSkin.PYTHON)
game.add_player(player_1, "Team GreedyEachTurn", location=StartingLocation.CENTER)
game.add_player(player_2, "Team AntColonies", location=StartingLocation.CENTER)
# Start the game
stats = game.start()
```

We visualize results using a pretty printer.

```python
# Show statistics
pprint.pprint(stats)
```
